## Завдання 1

$$
z1 = \frac{\sin(4\alpha)}{1+\cos(4\alpha)} \cdot \frac{\cos(2\alpha)}{1+\cos(2\alpha)}
$$

$$
z2 = \frac{1}{\tan\left(\frac{3}{2}\pi - \alpha\right)}
$$

```python
import math

alpha = 0.5  # можна змінити

z1 = (math.sin(4 * alpha) / (1 + math.cos(4 * alpha))) * \
     (math.cos(2 * alpha) / (1 + math.cos(2 * alpha)))

z2 = 1 / math.tan((3/2) * math.pi - alpha)

print("z1 =", z1)
print("z2 =", z2)
```

## Завдання 2

$$
\arctan(x) = -\frac{\pi}{2} + \sum_{k=0}^{n} \frac{(-1)^{k+1}}{(2k+1)x^{2k+1}}
$$

```python
import math

# параметри
x = 0.5
n = 20

# 1 спосіб (вбудована функція)
y1 = math.atan(x)

# 2 спосіб (ряд Тейлора для arctan)
def arctg(x, n):
    return sum(((-1)**k) * x**(2*k+1) / (2*k+1) for k in range(n))

y2 = arctg(x, n)

# результат
print("Через math.atan:", y1)
print("Через ряд:", y2)
print("Похибка:", abs(y1 - y2))
print("Кількість членів:", n)
```
